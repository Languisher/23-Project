```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
```

```python
import matplotlib.animation as animation
import IPython.display as display
```

```python
def enveloppe2D(points):

    hull = ConvexHull(points)


    plt.plot(points[:,0], points[:,1], 'o',color='red')

    
    for s in hull.simplices:
        plt.plot(points[s, 0], points[s, 1], 'k-')
        
  
    plt.title('Enveloppe Convexe')
    plt.xlabel('X')
    plt.ylabel('Y')

    
    plt.show()
```

```python
A=np.array([(0,0),(0,1),(2,3),(3,4),(4,5),(5,6)])
enveloppe2D(A)
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
def enveloppe3D(points):

    

    hull = ConvexHull(points)

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    
    ax.scatter(points[:,0], points[:,1], points[:,2],color='red')

    
    for s in hull.simplices:
        ax.plot(points[s, 0], points[s, 1], points[s, 2], 'k-')
        
    
    ax.set_title('Enveloppe Convexe')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Fonction d'animation
    def animate(i):
        ax.view_init(elev=10., azim=i)

    # Animate
    ani = animation.FuncAnimation(fig, animate, frames=360, interval=100)

    plt.show()
    return ani.save('animation.mp4')
```

```python
points = np.array([(0, 0, 0), (1, 3, 2), (2, 2, 1), (3, 5, 3), (4, 1, 2), (5, 6, 1)])
enveloppe3D(points)
```
